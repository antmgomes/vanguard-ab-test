```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_final_demo= pd.read_csv("https://drive.google.com/uc?id=1n-do3beXJIXcVdcUp8gAhIb4jbSjomWK")
df_final_exp_clients=pd.read_csv("https://drive.google.com/uc?id=19mSpzew88dgaSQKvpbS1Dt0L7h-WKk50")
df_web_data_1=pd.read_csv("https://drive.google.com/uc?id=1sTNwBd9Gr2_v3eNuAH-YbZ7Jh6YrKTPD")
df_web_data_2=pd.read_csv("https://drive.google.com/uc?id=1dUbosgeNvDPAHe-i761aF-YH20raALek")
df_final_demo
df_final_exp_clients
df_web_data_1

df_web=pd.merge (df_web_data_1,df_web_data_2, on="client_id",how="inner")
```

```python
df_web = df_web.drop_duplicates(subset="client_id", keep="first")
df_web
```

```python
df_final_demo["gendr"] = df_final_demo["gendr"].fillna(np.nan)
df_final_demo = df_final_demo.drop_duplicates(subset="client_id", keep="first")
df_final_demo["gendr"].replace("U",np.nan,inplace=True)
df_final_demo
```

```python
df_final_exp_clients
df_final_exp_clients = df_final_exp_clients.drop_duplicates(subset="client_id", keep="first")
df_final_exp_clients
```

```python
df_master = (
    df_web
    .merge(df_final_demo, on="client_id", how="inner")
    .merge(df_final_exp_clients, on="client_id", how="inner")
)
```

```python
df_master = df_web.merge(df_final_exp_clients, on = 'client_id', how = 'left').merge(df_final_demo, on = 'client_id', how = 'left')
```

```python
df_master['date_time_x'] = pd.to_datetime(df_master['date_time_x'], errors='coerce')
df_master['date_time_y'] = pd.to_datetime(df_master['date_time_y'], errors='coerce')
df_master.drop_duplicates(subset='client_id', inplace=True)
df_master["time_diff"]= df_master["date_time_y"]-df_master["date_time_x"]
df_master["time_diff"].describe
```

```python
df_master['session_duration'] = (df_master.groupby('client_id')['date_time_y'].transform('max') - df_master.groupby('client_id')['date_time_x'].transform('min'))
df_master['session_minutes'] = df_master['session_duration'].dt.total_seconds() / 60
df_master['session_minutes'].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df_master.boxplot(column='session_minutes', by='Variation', grid=False)
plt.title('Distribución del tiempo por diseño')
plt.suptitle('')
plt.xlabel('Variación')
plt.ylabel('Tiempo de sesión (minutos)')
plt.show()
```

```python
#not longer than a day 
df_clean = df_master[df_master['session_minutes'] < 1440]  
```

```python
#KPIs:
# 1. Complete rate
# 2. Time Spent per step -- per design (only  in the ones that are completed?)
# 3. Where they drop off?
# 4. % of drop off?
    #take out the out liers 
```

```python
df_master['process_step_y'].unique()
```

```python
#KPIs
#1. Completion Rate (% of clients who finished the online process - Higher = better experience)

#Confirm is the last step of the process - so it represents the clients who completed the process
last_step = 'confirm'

# Unique clients who reached the final step
completion = df_master[df_master['process_step_y'] == last_step].groupby('Variation')['client_id'].nunique()

# Total unique clients per group
total_clients = df_master.groupby('Variation')['client_id'].nunique()

# Completion rate (%)
completion_rate = (completion / total_clients) * 100
print(f"Completion Rate by Group: {completion_rate}")
```

```python
#2. Average time spent per step (how long clients take on each step - Lower time = smoother, easier process)

# Calculate time spent per row (step) in seconds
df_master['time_spent_sec'] = (df_master['date_time_y'] - df_master['date_time_x']).dt.total_seconds()

df_clean = df_master[df_master['time_spent_sec'] < 86400]
# Average time spent per step by group
avg_time_per_step = df_clean.groupby(['Variation', 'process_step_y'])['time_spent_sec'].mean().reset_index()

# Display in minutes
avg_time_per_step['time_spent_min'] = avg_time_per_step['time_spent_sec'] / 60
print(avg_time_per_step[['Variation', 'process_step_y', 'time_spent_min']].round())

#Variation - The design version in the A/B test: "Control" (old design) or "Test" (new design)
```

```python
# 3. Where they drop off?
step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']
step_counts = df_master.groupby(['Variation','process_step_x'])['client_id'].nunique().unstack('Variation').reindex(step_order)
step_counts
step_percentage = (step_counts / step_counts.iloc[0]) * 100
step_percentage
#in which step do we loss more people
#diff between start= start - step_1 .... 69%
# We loss 69% in between those steps

drop_off = step_percentage.diff(-1).abs()
drop_off
```

```python
drop_off.plot(kind='bar', figsize=(8,5), title='Drop-off por paso y variación')
plt.ylabel('% de usuarios perdidos respecto al paso siguiente')
plt.xlabel('Paso del proceso')
plt.show()
```

```python
#4. At which step in the process are most users quitting before completion?
max_step = df_master['process_step_x'].max()
max_step

df_master["completed"]= (df_master["process_step_x"]==max_step).astype(int)
df_master["error"]= (df_master["process_step_x"]<max_step).astype(int)
completation_rate= df_master.groupby("Variation")["completed"].mean()*100
error_rate = df_master.groupby('Variation')['error'].mean() * 100
```

```python
kpi_df = pd.concat([completation_rate, error_rate], axis=1)
kpi_df.columns = ['Completion Rate (%)', 'Error Rate (%)']
print(kpi_df)
```

```python
import matplotlib.pyplot as plt
```

```python
ax = kpi_df.plot(
    kind='bar',
    figsize=(8, 5),
    title='KPI Comparison by Variation',
)
plt.ylabel('%')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.legend(title='KPI')

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge')

plt.show()
```

Hypothesis 1 — Completion Rate

H0:
There is no significant difference in the completion rate between the Control and Test designs.

H1:
There is a significant difference in the completion rate between the Control and Test designs.

Hypothesis 2 — Average Time Spent per Step

H0:
There is no significant difference in the average time spent per step between the Control and Test designs.

H1:
There is a significant difference in the average time spent per step between the two designs.

Hypothesis 3 — Drop-off Rates

H0:
The proportion of users who abandon the process at each step is the same for both designs.

H1:
The proportion of users who abandon the process differs between designs.

Hypothesis 4 — Error Rate

H0:
There is no difference in the error rate (percentage of incomplete processes) between the Control and Test designs.

H1:
There is a difference in error rates between the two designs.
